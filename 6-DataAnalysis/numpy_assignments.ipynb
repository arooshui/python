```python
# 1. Create a NumPy array of shape (3, 3) representing a matrix. 
# Compute its determinant, inverse, and eigenvalues.
# 2. Create two NumPy arrays of shape (2, 3) and (3, 2). Perform 
# matrix multiplication on these arrays.
A = np.array([
    [2, 1, 3],
    [1, 0, 2],
    [4, 1, 8]
])
det = np.linalg.det(A)
inv = np.linalg.inv(A)
eigenvalues = np.linalg.eigvals(A)
print("Matrix A:")
print(A)
print("\nDeterminant:", det)
print("\nInverse:")
print(inv)
print("\nEigenvalues:")
print(eigenvalues)



a = np.array([
    [1, 2, 3],
    [4, 5, 6]
])   

B = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])   

result = np.dot(A, B)

print("Result of matrix multiplication:")
print(result)
```

```python
# 1. Create a structured array with fields 'name' (string), 'age' (integer), and
#  'weight' (float). Add some data and sort the array by age.
# 2. Create a structured array with fields 'x' and 'y' (both integers). 
# Add some data and compute the Euclidean distance between each pair of points.
import numpy as np
dtype = [('name', 'U10'), ('age', 'i4'), ('weight', 'f4')]
people = np.array([
    ('Aroosh', 22, 60.5),
    ('Ravi', 25, 70.2),
    ('Anil', 20, 55.0),
    ('John', 23, 68.4)
], dtype=dtype)
print("Before sorting:")
print(people)
sorted_people = np.sort(people, order='age')

print("\nAfter sorting by age:")
print(sorted_people)


dtype = [('x', 'i4'), ('y', 'i4')]

points = np.array([
    (1, 2),
    (4, 6),
    (8, 10),
    (3, 7)
], dtype=dtype)

print("Points:")
print(points)
distances = np.sqrt(points['x']**2 + points['y']**2)

print("\nEuclidean distances from origin:")
print(distances)
```

```python
# 1. Create a masked array of shape (4, 4) with random integers and mask the 
# elements greater than 10. Compute the sum of the unmasked elements.
# 2. Create a masked array of shape (3, 3) with random integers and mask the
# diagonal elements. Replace the masked elements with the mean of the unmasked 
# elements.
import numpy as np
a = np.random.randint(1, 20, size=(4,4))
print("Original array:")
print(a)
masked_arr = np.ma.masked_where(a > 10, a)

print("\nMasked array:")
print(masked_arr)
print("\nSum of unmasked elements:", masked_arr.sum())



b = np.random.randint(1, 20, size=(3,3))
print("Original array:")
print(b)
mask = np.eye(3, dtype=bool)
masked_b = np.ma.array(b, mask=mask)

print("\nMasked array (diagonal masked):")
print(masked_b)

# Compute mean of unmasked values
mean_val = masked_b.mean()

# Replace masked elements with mean
filled_array = masked_b.filled(mean_val)

print("\nFinal array after replacing masked with mean:")
print(filled_array)

```
