```python
students = [
    {"name": "Aroosh", "score": 85},
    {"name": "Rahul", "score": 92},
    {"name": "Sneha", "score": 78},
    {"name": "Kiran", "score": 88}
]
n = len(students)
for i in range(n):
    for j in range(0,n-i-1):
        if students[j]["score"] <= students[j+1]["score"]:
            students[j]["score"] , students[j+1]["score"] = students[j+1]["score"],students[j]["score"]
print(students)
```

```python
def transpose(mat):
    matrix=[]
    for i in range(3):
        row=[]
        for j in range(3):
            row.append(mat[j][i])
        matrix.append(row)
    return matrix
matrix = []
for i in range(3):
    row=[]
    for j in range(3):
        value = int(input(f"Enter element{i},{j}"))
        row.append(value)
    matrix.append(row)
for row in matrix:
    print(row)
trans = transpose(matrix)
print()
for row in trans:
    print(row)
```
