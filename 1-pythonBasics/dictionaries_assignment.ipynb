```python
def func(i):
    li=[]
    for j in range(1,6):
        li.append(i*j)
    return li
du={}
for i in range(1,6):
    du[i] = func(i)
print(du)
```

```python
def func1(i):
    tup = (i,i**2)
    return tup
du={}
for i in range(1,6):
    du[i] = func1(i)
print(du)
```

```python
du={}
for i in range(1,6):
    du[i] = i**2
li=[]
for key,value in du.items():
    li.append((key,value))
print(li)
```

```python
d1 = {x:x**2 for x in range(1,6)}
d2={}
for key,values in d1.items():
    d2[values]=key
print(d2)
```
