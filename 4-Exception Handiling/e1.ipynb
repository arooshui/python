```python
try:
    n = int(input("Enter a number"))
    c  = 10/n
    print(c)
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter any another numer")
except Exception as e1:
    print(e1)
```

```python
try:
    n = int(input("Enter a number"))
    c  = 10/n
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter any another numer")
except Exception as e1:
    print(e1)
else:
    print(c)
finally:
    print("Execution Completed")
```
