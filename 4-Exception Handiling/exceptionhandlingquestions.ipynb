```python
# Write a function that reads the contents of a file named `data.txt`.
# Use try, except, and finally blocks to handle file not found errors and 
# ensure the file is properly closed.
try:
    file=open('a.txt','r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File doesnt exist")
except Exception as ex:
    print(ex)
finally:
    if 'file' in locals() or not file.closed:
        file.close()
        print('file close')
```

```python
# Write a function that prompts the user toenter an integer. 
# Use try, except, and finally blocks to handle ValueError if the user enters 
# a non-integer value and print an appropriate message.

def get_integer():
    try:
        num = int(input("Enter an integer: "))
        print("You entered:", num)

    except ValueError:
        print("Invalid input! Please enter only an integer.")

    finally:
        print("Program execution completed.")
        
get_integer()
```

```python
# Write a function that performs nested exception handling. It should first
# attempt to convert a string to an integer, and then attempt to divide by that
# integer. Use nested try, except, and finally blocks to handle ValueError and
#  ZeroDivisionEcrror and print appropriate messages.
def div(n,b):
    # c=1
    try:
        a = int(n)
        b = int(b)
        try:
            c=a/b
        except ZeroDivisionError:
            print("Dont enter zero in denominator")
        else:
            print(c)
        finally:
            print("Division attempt finished")
    except ValueError:
        print("Enter a valid number")
    finally:
        print("program execution finished")
    
    

n=input("enter number1")
b=input("enter number 2")
div(n,b)
```

```python
# Write a function that takes a list and an index as input and returns the 
# element at the given index. Use try, except, and finally blocks to handle
#  IndexError if the index is out of range and print an appropriate message.
def ind(li,n):
    try:
        val = li[n]
    except IndexError:
        print("Index is out of range")
    else:
        print(val)
li=[1,2,3,4,5]
print(ind(li,3))
print(ind(li,6))
```

```python
# Write a function that attempts to parse a JSON string. Use try, except, and 
# finally blocks to handle JSONDecodeError if the string is not a valid JSON and
#  print an appropriate message.
import json
def parse_json(text):
    try:
        data = json.loads(text)
        print("Valid JSON:", data)

    except json.JSONDecodeError:
        print("Invalid JSON format")

    finally:
        print("Parsing attempt finished")


parse_json('{"name": "Aru", "age": 21}')
parse_json('{name: Aru, age: 21}')

```

```python
# Define a custom exception named `NegativeNumberError`. Write a function that
# raises this exception if a negative number is encountered in a list. Use try,
# except, and finally blocks to handle the custom exception and print an
# appropriate message.

class NegativeNumberError(Exception):
    pass
def check_numbers(li):
    try:
        for num in li:
            if num < 0:
                raise NegativeNumberError("Negative number found: " + str(num))
        print("All numbers are valid")

    except NegativeNumberError as e:
        print("Custom Exception:", e)

    finally:
        print("Checking completed")
check_numbers([1,1,2,3,-1])

```

```python
# Write a function that calls another function which may raise an exception. 
# Use try, except, and finally blocks to handle the exception and print an 
# appropriate message

def risky(x):
   return 10/x

def caller(x):
    try:
        result = risky(x)
        print("Result:", result)

    except ZeroDivisionError as e:
        print("Error occurred:", e)

    finally:
        print("Execution completed")
caller(0)
```

```python
# Write a function that takes a list of strings and converts them to integers.
# Use try, except, and finally blocks to handle ValueError if a string cannot be
#  converted and print an appropriate message.
def fun(li):
    l2=[]
    try:
        for i  in range(len(li)):
            l2.append(int(li[i]))
        
    except ValueError:
        print("cant convert it into integer")
    else:
        print(li)
    finally:
        print("Exec completed")
        print(l2)
li=['1','2','3','add']
fun(li)
```

```python
# Write a function that uses a list comprehension to convert a list of strings to
# integers. Use try, except, and finally blocks within the list comprehension to
#  handle ValueError and print an appropriate message.
def convert(x):
    try:
        return int(x)
    except ValueError:
        print(f"Cannot convert '{x}' to integer")
    finally:
        print("exec completed")

def convert_list(lst):
    return [convert(i) for i in lst]

print(convert_list(["10", "20", "abc", "40"]))
```

```python
# Write a function that attempts to write a list of strings to a file.
# Use try, except, and finally blocks to handle IOError and ensure the file is 
# properly closed.
def write_to_file(data):
    file = None
    try:
        file = open("output.txt", "w")
        for line in data:
            file.write(line + "\n")
        print("Data written successfully")
    except IOError:
        print("Error: Unable to write to file")
    finally:
        if file:
            file.close()
            print("File closed")

write_to_file(["Hello", "Python", "Exception Handling"])
```
