File Handling

```python
#reading a whole file 
with open('ex.txt','r') as file:
    content = file.read()
    print(content)
```

```python
#read a file line by line
with open('ex.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
#writing file with overwritting
with open('ex.txt','w') as file:
    file.write('hello guys \n')
    file.write('hello boys and girls \n')
```

```python
#writing file without overwritting
with open('ex.txt','a') as file:
    file.write('hello ladies and gentleman \n')
```

```python
#writing list of lines into file
lines=['hiii \n','byeeee \n' ,'see youuu \n']
with open('ex.txt','a') as file:
    file.writelines(lines)
```

```python
#read the contents of a file from a source file and write it to the destination file
with open('ex.txt','r') as file:
    content = file.read()
with open('e2.txt','w') as file:
    file.write(content)
```

```python
#read a text file count the number of lines,words,characters
def count(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count,word_count,char_count
path = 'ex.txt'
lines,words,char = count(path)
print(f"lines:{lines} , words:{words} , characters:{char}")
```
