```python
def count(path):
    with open(path,'r') as file:
        content  = file.read()
    return len(content.split())
path = 'output.txt'
words = count(path)
print(words)
```

```python
#Write a function that finds and replaces all occurrences of a given word in a file named `data.txt` with another word.
def replace(old,new,path):
    with open(path,'r') as file:
        content = file.read()
    content = content.replace(old,new)

    with open(path,'w') as file:
        file.write(content)

path='e3.txt'
old='hello'
new='bello'
with open(path,'r') as file:
    content = file.read()
    print(content)
replace(old,new,path)
with open(path,'r') as file:
    content = file.read()
    print(content)
```

```python
def read(path):
    with open(path,'r') as file:
        content = file.readlines()
        for i in content:
            print(i[::-1],end="")
path = 'ex.txt'
read(path)
```

```python
#read a text file count the number of lines,words,characters
def count(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count,word_count,char_count
path = 'ex.txt'
lines,words,char = count(path)
print(f"lines:{lines} , words:{words} , characters:{char}")
```

```python
#Write a function that merges the contents of multiple files into a single file named `merged.txt`.
def merge(p1,p2,p3):
    with open(p1,'r') as file:
        c1 = file.read()
    with open(p2,'r') as file:
        c2 = file.read()
    with open(p3,'w') as file:
        file.write(c1+c2)
p1 = 'ex.txt'
p2 = 'e2.txt'
p3 = 'merge.txt'
merge(p1,p2,p3)
with open(p3,'r') as file:
    c2 = file.read()
    print(c2.strip())
```

```python
#Write a function that splits a large file named `large.txt` into smaller files of 100 lines each.
def split_file(file):
    with open(file, 'r') as file:
        lines = file.readlines()

    part = 1
    for i in range(0, len(lines), 100):
        with open(f'part{part}.txt', 'w') as new_file:
            new_file.writelines(lines[i:i+100])
        part += 1

file = 'large.txt'
split_file(file)
```

```python
#Write a function that reads a binary file named `image.bin` and writes its contents to another binary file named `copy_image.bin`.
def copy_binary_file():
    with open("image.bin", "rb") as source:
        data = source.read()

    with open("copy_image.bin", "wb") as destination:
        destination.write(data)

    print("Binary file copied successfully!")

copy_binary_file()
with open('copy_image.bin','rb') as file:
    print(file.read())
```

```python
#Write a function that reads a CSV file named `data.csv` and prints its contents as a list of dictionaries.
import csv
def read():
    with open('data.csv','r') as file:
        reader = csv.DictReader(file)
        data = list(reader)

    print(data)

read()
```

```python
#Write a function that reads a JSON file named `data.json` and prints its contents as a Python dictionary.
import json
def read():
    with open('data.json','r') as file:
        data = json.load(file)
    print(data)
read()
```

```python
# Write a function that attempts to read a file named protected.txt and 
# handles any permission errors gracefully by printing an error message.
def read():
    try:
        with open("e2.txt", "r") as file:
            content = file.read()
            print(content)
    except PermissionError:
        print("Error: You do not have permission to read this file.")
    except FileNotFoundError:
        print("Error: The file 'protected.txt' was not found.")

read()
```
