```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
n=int(input("Enter the number"))
print(fib(n))
```

```python
def fun(a,b={}):
    d={}
    d[a] = len(b)+1
    return d
print(fun("apple"))
print(fun("ball"))
cust = {"he":1}
print(fun("app",cust))

```

```python
def keyword(**kwargs):
    d={}
    for key,value in kwargs.items():
        d[key]=value
    return d
print(keyword(a=5,b=10,aru=122,d=24))
print(keyword(a1=5,b11=10,d=24))
print(keyword(a=4,b=3))
```

```python
def squ(x):
    return x**2
def dob(x):
    return  x*2
def cub(x):
    return x**3
def callbackfunc(callback,li):
    lis=[]
    for n in li:
        lis.append(callback(n))
    return lis
li=[1,2,3]
print(callbackfunc(squ,li))
print(callbackfunc(dob,li))
print(callbackfunc(cub,li))
```

```python
def outer():
    def inner(n):
        return n * n
    return inner

squarefunc = outer()
print(squarefunc(2))
print(squarefunc(5))
print(squarefunc(10))
```

```python
def eve(n):
    return n%2==0
def pos(n):
    return n>0
def squ(n):
    return n**2
def cub(n):
    return n**3
def call(filt,mapp,li):
    fil = []
    for i in li:
        if filt(i):
            fil.append(i)
    mapped=[]
    for i in fil:
        mapped.append(mapp(i))
    return mapped

li = [1,2,3,4,5]
print(call(eve,squ,li))
print(call(pos,cub,li))
```

```python
def squ(n):
    return n**2
def cub(n):
    return n**3
def call(f,g,n):
    return f(g(n))
print(call(cub,squ,3))
```

```python
from functools import partial
def mul(n,m):
    return n*m
into2 = partial(mul,2)
print(into2(5))
```

```python
def mul(a):
    def in1(b):
        def in2(c):
            return a*b*c
        return in2
    return in1
print(mul(5)(2)(3))
```

```python
def fun(li):
    l1=[]
    l2=[]
    l3=[]
    for i in li:
        if type(i)==int:
            l1.append(i)
        elif type(i)==float:
            l2.append(i)
        elif type(i)==str:
            l3.append(i)
    return  l1,l2,l3
li=[1,2.0,"abc",5,"ax",6.0,3,"abb",4,5.7]
print(fun(li))
```
