```python
#Data serialization
import json
data={'a':1, "b":1}
js = json.dumps(data)
print(type(js))
print(js)
dic = json.loads(js)
print(type(dic))
print(dic)
```

```python
#csv
import csv
with open('ex.csv',mode='w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['aru','abc'])
    writer.writerow(['varo','cde'])
with open('ex.csv',mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
