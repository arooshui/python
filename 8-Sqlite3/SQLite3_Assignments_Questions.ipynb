```python
import sqlite3
```

```python
connection = sqlite3.connect('test1.db')
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''
create table if not exists employees(
    id integer primary key,
    name text,
    age integer,
    department text)
 ''')
connection.commit()
```

```python
cursor.execute('''
insert into employees(id,name,age,department)
            values
                (1,"aru",20,"aiml") 
''')
connection.commit()
```

```python
cursor.execute('select * from employees')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
employees_data = [
    (2,"aaa",21,"ds"),
    (3,"bbb",22,"cse"),
    (4,"ccc",33,"csd"),
    (5,"ddd",444,"cssd")
]

cursor.executemany('''insert into employees(id,name,age,department)
               values(?,?,?,?)''',employees_data)
connection.commit()
```
